#### 混合正規分布の生成

```python
#ライブラリのインポート
import numpy as np
import matplotlib.pyplot as plt


### Generate data

mu1 = np.array([0,0])
mu2 = np.array([1,1])
Sigma = np.array([[2,-1],[-1,2]])

N1 = N2 = 30

np.random.seed(0)
# Label1
xx1 = np.random.multivariate_normal(mu1,Sigma,N1)
yy1 = np.array([1]*N1)

#Label2
xx2 = np.random.multivariate_normal(mu2,Sigma,N2)
yy2 = np.array([-1]*N2)

xx = np.concatenate([xx1,xx2])
yy = np.concatenate([yy1,yy2])
```

```python
# 混合正規分布のプロット

plt.xlabel("x2")
plt.ylabel("x1")
plt.scatter(xx1[:,1],xx1[:,0],marker="o",alpha=0.5)
plt.scatter(xx2[:,1],xx2[:,0],marker="^",alpha=0.5)
plt.show()
```

```python
## NNによる分類用パッケージ

from sklearn.neural_network import MLPClassifier as MLP
from sklearn.metrics import accuracy_score

## NNの計算
clf = MLP(solver='lbfgs', alpha=1e-5, hidden_layer_sizes=(1), random_state=1) #モデルの型を指定
clf.fit(xx, yy)
clf.predict(xx) #訓練データに関する予測値
clf.predict_proba(xx) #訓練データに関する予測確率
print(type(clf.predict_proba(xx)))
print(np.mean(clf.predict_proba(xx), axis=0))
```

### 3-1

```python
## ネットワーク描画 
from draw_neural_net3 import draw_neural_net3

# 描画用の領域を作成
plt.clf()
fig1 = plt.figure(figsize=(6, 6))
ax = fig1.gca()
ax.axis('off')

# MLPCalssifierの出力から中間層の層数とユニット数を取得
nn_layer_sizes = [np.ndim(xx)]
print(nn_layer_sizes)
for num in range(len(clf.intercepts_)-1):
  nn_layer_sizes = np.append(nn_layer_sizes, clf.coefs_[num].shape[1])
nn_layer_sizes = np.append(nn_layer_sizes,np.ndim(yy))

print(nn_layer_sizes)
print("train_error is "+str(1-accuracy_score(clf.predict(xx),yy)))

# ネットワーク出力
draw_neural_net3(ax, .1, .9, .1, .9, nn_layer_sizes,
                 clf.coefs_, clf.intercepts_, clf.n_iter_, clf.loss_)

plt.savefig("14-3-1.pdf")
# plt.savefig('fig1.png')   # ファイルに出力するときに用いる
```

```python
import pandas as pd
import urllib.request
url = 'http://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-white.csv'
urllib.request.urlretrieve(url, 'winequality-white.csv')
data = pd.read_csv('winequality-white.csv', sep=';') 

import numpy as np
import pandas as pd
from sklearn.neural_network import MLPClassifier as MLP

from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split, GridSearchCV


train, test = train_test_split(data, train_size=0.2, stratify=data.quality, random_state=1)
train0, test0 = train_test_split(data, train_size=0.2, random_state=1)
```

### 4-1

```python
freq_all = pd.crosstab(index=data.quality, columns="count")
#print(freq_all)

print()
print("train's length : "+str(len(train)))
print("test's length : "+str(len(test)))
print("train0's length : "+str(len(train0)))
print("test0's length : "+str(len(test0)))
print(len(train)/(len(train)+len(test)))
```

```python
freq_train = pd.crosstab(index=train.quality, columns="count")
print(freq_train)
freq_train0 = pd.crosstab(index=train0.quality, columns="count")
print(freq_train0)
```

```python
freq_test = pd.crosstab(index=test.quality, columns="count")
print(freq_test)
freq_test0 = pd.crosstab(index=test0.quality, columns="count")
print(freq_test0)
```

### 4-2

```python
train_bin = train
train_bin.quality = (train.quality<=5)

test_bin = test
test_bin.quality = (test.quality<=5)

train_X = train_bin.drop('quality',axis=1)
train_y = train_bin.quality

test_X = test_bin.drop('quality',axis=1)
test_y = test_bin.quality


param_grid ={'hidden_layer_sizes':[(1,),(3,),(5,)],'alpha':[0.00001,0.0001,0.001]} 
grid_search = GridSearchCV(MLP(random_state=1), param_grid, cv=10)
grid_search.fit(train_X, train_y)
```

```python
print('Test score: {}'.format(grid_search.score(test_X, test_y))) 
print('Best parameters: {}'.format(grid_search.best_params_))
```

### 4-3

```python
pred_test_y = grid_search.best_estimator_.predict(test_X)
conf_mat = confusion_matrix(test_y,pred_test_y)

print(conf_mat)
print((conf_mat[1][1]+conf_mat[0][0])/np.sum(conf_mat))
print(accuracy_score(test_y,pred_test_y))
```
