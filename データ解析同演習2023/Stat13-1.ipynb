# 問題3

```python
#ライブラリのインポート
import numpy as np
import matplotlib.pyplot as plt

### Generate data

mu1 = np.array([0,0])
mu2 = np.array([1,1])
Sigma = np.array([[2,-1],[-1,2]])

N1 = N2 = 30

np.random.seed(0)
```

### 問題3-1

```python
# Label1
np.random.seed(0)
xx1 = np.random.multivariate_normal(mu1,Sigma,N1)
yy1 = np.array([1]*N1)
#Label2
xx2 = np.random.multivariate_normal(mu2,Sigma,N2)
yy2 = np.array([-1]*N2)

xx = np.concatenate([xx1,xx2])
yy = np.concatenate([yy1,yy2])

plt.figure(figsize=(8.3, 5.8))

plt.xlabel("x2")
plt.ylabel("x1")
plt.scatter(xx1[:,1],xx1[:,0],marker="o",alpha=0.5)
plt.scatter(xx2[:,1],xx2[:,0],marker="^",alpha=0.5)
plt.show()
```

```python
### Linear SVM
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score

c=100
svm = LinearSVC(C=c)
svm.fit(xx, yy)
yy_pred = svm.predict(xx)
print("C="+ str(c)+": train_error is "+str(1-accuracy_score(yy_pred,yy)))
```

```python
#分類関数の図示
x = np.arange(-5,5,0.1)
y = np.arange(-5,5,0.1)
X,Y = np.meshgrid(x, y)
Z = svm.decision_function(np.c_[X.ravel(),Y.ravel()])
Z = Z.reshape(X.shape)

plt.figure(figsize=(8.3, 5.8))

plt.xlabel("x2")
plt.ylabel("x1")
plt.scatter(xx1[:,1],xx1[:,0],marker="o",alpha=0.5)
plt.scatter(xx2[:,1],xx2[:,0],marker="^",alpha=0.5)
cont = plt.contour(X, Y, Z,colors=['r', 'g'])
cont.clabel(fmt='%1.1f', fontsize=14)
plt.rcParams['font.family'] = 'Yu Gothic'
plt.title("3-1. C=1での線形SVM")
# PDFファイルとして保存
plt.savefig("13-3-1.pdf")
plt.show()
```

```python
## データの読み込み
star_train = pd.read_csv("C:\python\データ解析同演習2023\star_training200.csv") #パスは自分のを使用
print(star_train.head())

X_train = star_train.drop("main.seq",axis=1) #データの属性
y_train = star_train["main.seq"] #データのラベル
```

### 問題4-1

```python
#ガウスカーネル(σ=1)
svm_g = SVC(kernel="rbf",C=1,gamma=1) #gamma(分散)の値を変える
svm_g.fit(X_train,y_train)
plt.clf()
plt.figure(figsize=(8.3, 5.8))

#等高線の図示
x_min, x_max = X_train["amag"].min()-1, X_train["amag"].max()+1
y_min, y_max = X_train["class"].min()-1, X_train["class"].max()+1
X, Y = np.meshgrid(np.arange(x_min, x_max, 0.1), np.arange(y_min, y_max, 0.1))
Z = svm_g.decision_function(np.c_[X.ravel(), Y.ravel()])
Z = Z.reshape(X.shape)

fig, ax = plt.subplots(figsize=(8.3, 5.8))
for name, group in star_train.groupby("main.seq"):
    ax.plot(group["amag"], group["class"], marker='o', linestyle='', ms=8, label=name)
    ax.legend()

# cont = ax.contour(X, Y, Z,colors=['r', 'g'],levels=10)
# cont.clabel(fmt='%1.1f', fontsize=14)


ax.contourf(X, Y, Z, colors=['blue', 'red'], levels=0, alpha=0.2)
cont = ax.contour(X, Y, Z, colors=['r', 'g'], levels=10)
cont.clabel(fmt='%1.1f', fontsize=14)


plt.xlabel("amag")
plt.ylabel("class")
plt.rcParams['font.family'] = 'Yu Gothic'
plt.title("4-1. $C=1,\sigma=1$" "のガウスカーネルを用いたSVM")
plt.savefig("13-4-1-1.pdf")
plt.show()

#サポートベクターの個数
print("support_vectors: "+str(svm_g.n_support_))
```
