```python
import numpy as np
import math
import matplotlib.pyplot as plt

    
#plotに使用する配列の保存先
x_list =[]
y_list_1 = []
y_list_2 = []
y_list_3 = []

#plotに使用する配列の保存
x = 0.0
for i in range(1001):
    x_list.append(x)
    y_list_1.append(math.sin(x))
    y_list_2.append(math.sin(10 * x))
    y_list_3.append(math.sin(100 * x))
    x += 0.001
    
#plot
plt.plot(x_list, y_list_1, label = "y = sin(x)")
plt.plot(x_list, y_list_2, label = "y = sin(10x)")
plt.plot(x_list, y_list_3, label = "y = sin(100x)")
plt.title("plot of sin(wx)")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.savefig("sin(wx).png")
plt.show()


```

```python
import numpy as np
import math
import matplotlib.pyplot as plt

    
#plotに使用する配列の保存先
x_list =[]
y_list = []

#plotに使用する配列の保存
x = 0.001
for i in range(10001):
    x_list.append(x)
    y_list.append((x ** 4) * (1 - math.sin(x) / x) / 2)
    x += 0.001
    
#plot
plt.plot(x_list, y_list, label = r"$y = \frac{\omega ^4}{2} (1 - sinc (2 \omega))$")
# plt.title(r"plot of  $\frac{\omega ^4}{2} (1 - sinc (2 \omega))$")
plt.xlabel(r"$\omega$")
plt.ylabel("y")
plt.legend()
plt.savefig("wiggly_1.png")
plt.show()
```

```python
import numpy as np
import math
import matplotlib.pyplot as plt

    
#plotに使用する配列の保存先
x_list =[]
y_list = []

#plotに使用する配列の保存
w = 0
x = -10.0
for i in range(20001):
    x_list.append(x)
    y_list.append((max(w,x) * min(w,x) ** 2) / 2 - min(x,w) ** 3 / 6)
    x += 0.001
    
#plot
plt.plot(x_list, y_list, label = r"$y = \kappa (w, x)$")
# plt.title(r"plot of  $\frac{\omega ^4}{2} (1 - sinc (2 \omega))$")
plt.xlabel(r"$x$")
plt.ylabel(r"$y$")
plt.legend()
plt.savefig("cubic_spline.png")
plt.show()
```

```python
import numpy as np
import math
import matplotlib.pyplot as plt

    
#plotに使用する配列の保存先
x_list =[]
y_list = []


def normal(x,a,mu):
    y = np.exp(-0.5 * (x - mu)* (x - mu) / a) / np.sqrt(2.0 * np.pi * a)
    return y

#plotに使用する配列の保存
w = 0
x = -10.0
for i in range(20001):
    x_list.append(x)
    y_list.append(normal(x,2,-5))
    x += 0.001
    
#plot
plt.plot(x_list, y_list)
# plt.title(r"plot of  $\frac{\omega ^4}{2} (1 - sinc (2 \omega))$")
plt.xlabel(r"$x$")
plt.ylabel("prob")
plt.legend()
plt.savefig("normal_dist1.png")
plt.show()
```
