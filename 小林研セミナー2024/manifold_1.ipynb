```python
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
ax = fig.add_subplot(projection='3d')

# パラメータ t, u のメッシュグリッド作成
t = np.linspace(0.0, 2 * np.pi, 1000)
u = np.linspace(0.0, np.pi, 1000)
t, u = np.meshgrid(t, u)

# 球面の座標を計算
a = 1  # 球の半径
x = a * np.cos(u) * np.cos(t)
y = a * np.sin(u) * np.cos(t)
z = a * np.sin(t)

# 条件に応じた色設定
colors = np.full(x.shape, '#64adff', dtype=object)  # デフォルト青
mask = x**2 + y**2 < 0.0
colors[mask] = 'red'  # 条件を満たす部分を赤に

# 球面を描画
ax.plot_surface(x, y, z, facecolors=colors, edgecolor='none', rstride=1, cstride=1)

# 軸を非表示にする
ax.set_axis_off()

# グラフを表示
plt.savefig("sphere_colored.png")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D  # 明示的にインポート（場合によっては不要）
from matplotlib.colors import ListedColormap

# custom = ListedColormap(['#64adff'], name="custom")

# 図の作成
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

# t, u の解像度を適度に設定
t_vals = np.linspace(0, 2 * np.pi, 400)
u_vals = np.linspace(0, 2 * np.pi, 400)
t, u = np.meshgrid(t_vals, u_vals)

def plot_surface2(ax, a):
    # 楕円体のパラメトリック方程式
    # ※この式では、x^2+y^2+(z/2)^2 = 1 となる
    x = a * np.cos(u) * np.cos(t)
    y = a * np.sin(u) * np.cos(t)
    z = 2 * a * np.sin(t)
    
    # 冬のカラーマップ（winter）を用いてzの値から色を決定
    norm = plt.Normalize(z.min(), z.max())
    facecolors = plt.cm.winter(norm(z/10.0))
    
    # 条件: (x^2 + y^2 < 0.5) かつ (z > 0) の部分を赤色に変更
    mask = (x**2 + y**2 < 0.7) & (z > 0)
    facecolors[mask] = [1, 0, 0, 1]  # 赤 (RGBA)
    
    # 面の描画
    ax.plot_surface(x, y, z,facecolors=facecolors,rstride=1, cstride=1, edgecolor = 'none', shade = False)

# 球面（楕円体）を描画
plot_surface2(ax, 1)

# 軸を非表示
ax.set_xticks([])
ax.set_yticks([])
ax.set_zticks([])
ax.set_box_aspect([1,1,1])  # アスペクト比を均等に設定
ax.set_axis_off()

# 保存する場合は余白をなくして保存
plt.savefig("spere_07.png", bbox_inches='tight', pad_inches=0)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import math
from mpl_toolkits.mplot3d import Axes3D  # 明示的にインポート（場合によっては不要）
from matplotlib.colors import ListedColormap
import matplotlib.colors as mcolors  # RGBA変換に利用

# custom = ListedColormap(['#64adff'], name="custom")

# 図の作成
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

# t, u の解像度を適度に設定
x_vals = np.linspace(-1, 1, 200)
y_vals = np.linspace(-1, 1, 200)
x, y = np.meshgrid(x_vals, y_vals)

def plot_surface2(ax, a):
    z = 0 * x
    
    # デフォルト色：'#0083bf' をRGBAに変換
    default_color = mcolors.to_rgba('#008dcf')
    # x, y の形状に合わせたRGBAカラー配列を作成
    facecolors = np.full(x.shape + (4,), default_color)
    
    # 条件: (x^2 + y^2 < 0.5) かつ (z > 0) の部分を赤色に変更
    rad = math.asin(0.1)
    mask = (x**2 + y**2 < rad) 
    facecolors[mask] = [1, 0, 0, 1]  # 赤 (RGBA)
    
    # 面の描画
    ax.plot_surface(x, y, z,facecolors=facecolors,rstride=1, cstride=1, edgecolor = 'none', shade = False)

# 球面（楕円体）を描画
plot_surface2(ax, 1)

# 軸を非表示
ax.set_xticks([])
ax.set_yticks([])
ax.set_zticks([])
ax.set_box_aspect([1,1,1])  # アスペクト比を均等に設定
ax.set_axis_off()

# 保存する場合は余白をなくして保存
plt.savefig("plane_01.png", bbox_inches='tight', pad_inches=0)
plt.show()
```
