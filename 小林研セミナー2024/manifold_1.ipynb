```python
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
ax = fig.add_subplot(projection='3d')

x = y = np.arange(-7, 7, 0.01)
X, Y = np.meshgrid(x, y)
mu_x = 1
mu_y = 1
sigma_x = 8
sigma_y = 8

z = np.exp((-(X-mu_x)**2 / sigma_x- (Y-mu_y)**2 / sigma_y)) / np.sqrt(np.pi * sigma_x + sigma_y) 

ax.plot_surface(X,Y,z, color='#64adff')
plt.savefig("normal_1.png")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import math
from mpl_toolkits.mplot3d import Axes3D  # 
from matplotlib.colors import ListedColormap
import matplotlib.colors as mcolors  # RGBA

# custom = ListedColormap(['#64adff'], name="custom")

# 
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

# t, u 
x_vals = np.linspace(-1, 1, 200)
y_vals = np.linspace(-1, 1, 200)
x, y = np.meshgrid(x_vals, y_vals)

def plot_surface2(ax, a):
    z = 0 * x
    
    # '#0083bf' RGBA
    default_color = mcolors.to_rgba('#008dcf')
    # x, y RGBA
    facecolors = np.full(x.shape + (4,), default_color)
    
    # : (x^2 + y^2 < 0.5)  (z > 0) 
    rad = math.asin(0.1)
    mask = (x**2 + y**2 < rad) 
    facecolors[mask] = [1, 0, 0, 1]  #  (RGBA)
    
    # 
    ax.plot_surface(x, y, z,facecolors=facecolors,rstride=1, cstride=1, edgecolor = 'none', shade = False)

# 
plot_surface2(ax, 1)

# 
ax.set_xticks([])
ax.set_yticks([])
ax.set_zticks([])
ax.set_box_aspect([1,1,1])  # 
ax.set_axis_off()

# 
plt.savefig("plane_01.png", bbox_inches='tight', pad_inches=0)
plt.show()
```
