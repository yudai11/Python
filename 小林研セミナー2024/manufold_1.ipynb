```python
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
ax = fig.add_subplot(projection='3d')

x = y = np.arange(-1, 1, 0.01)
X, Y = np.meshgrid(x, y)

z = - (X**2 + Y**2) 

ax.plot_surface(X,Y,z, cmap="winter")
plt.savefig("mnf_1.png")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
ax = fig.add_subplot(projection='3d')

x = y = np.arange(-0.1, 0.1, 0.01)
X, Y = np.meshgrid(x, y)

z = Y * 0

ax.plot_surface(X,Y,z, color ="red")
plt.savefig("mnf_4.png")
plt.show()
```
