```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.random import rand
np.random.seed(1203)

fig = plt.figure(figsize=(10,8)) #図のサイズを決定
ax = fig.add_subplot(projection='3d') #3d空間の作成

x = y = np.arange(-5, 5, 1)

# x = [1,2]
# y = [3,4]

x2,y2 = np.meshgrid(x,y) #格子状の二次元配列作成（後ほど補足説明あり）

x3 = x2.ravel() #多次元配列を一次元に変更（後ほど補足説明あり）
y3 = y2.ravel() #多次元配列を一次元に変更
n = len(x3)
height = rand(n) #各棒グラフの高さ設定
height = height / sum(height)

bottom = np.zeros_like(height) #1のみで構成された配列を作成。要素数はheightと同じ
width = 1 #各棒グラフの横幅設定
depth = 1 #各棒グラフの奥行き設定

#各軸に対するラベルの作成
ax.set_xlabel('x') 
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.bar3d(x3,y3,bottom, width, depth, height) #3D棒グラフの生成
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.random import rand

np.random.seed(1)

fig = plt.figure(figsize=(10, 8))  # 図のサイズを決定
ax = fig.add_subplot(projection='3d')  # 3D空間の作成

# X, Y 軸の値を定義
x = y = np.arange(-5, 5, 1)
x2, y2 = np.meshgrid(x, y)  # 格子状の二次元配列を作成

# 一次元配列に変換（棒グラフの x, y の座標）
x3 = x2.ravel()
y3 = y2.ravel()

n = len(x3)  # 格子点の数

# 各棒グラフの高さをランダムに生成
height = rand(n)  # 長さ n のランダム配列
height = height / sum(height)

bottom = np.zeros_like(height)  # 各棒グラフの z 座標の開始位置
width = depth = 1  # 各棒グラフの横幅と奥行き設定

# 各軸に対するラベルを設定
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

# 3D棒グラフの生成
ax.bar3d(x3, y3, bottom, width, depth, height, shade=True, color='#9ddf6e')

# グラフの表示
plt.savefig("3dbarplt_1.png")
plt.show()
```
