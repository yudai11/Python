```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.random import rand
np.random.seed(1203)

fig = plt.figure(figsize=(10,8)) #
ax = fig.add_subplot(projection='3d') #3d

x = y = np.arange(-5, 5, 1)

# x = [1,2]
# y = [3,4]

x2,y2 = np.meshgrid(x,y) #

x3 = x2.ravel() #
y3 = y2.ravel() #
n = len(x3)
height = rand(n) #
height = height / sum(height)

bottom = np.zeros_like(height) #1height
width = 1 #
depth = 1 #

#
ax.set_xlabel('x') 
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.bar3d(x3,y3,bottom, width, depth, height) #3D
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.random import rand

np.random.seed(1)

fig = plt.figure(figsize=(10, 8))  # 
ax = fig.add_subplot(projection='3d')  # 3D

# X, Y 
x = y = np.arange(-5, 5, 1)
x2, y2 = np.meshgrid(x, y)  # 

#  x, y 
x3 = x2.ravel()
y3 = y2.ravel()

n = len(x3)  # 

# 
height = rand(n)  #  n 
height = height / sum(height)

bottom = np.zeros_like(height)  #  z 
width = depth = 1  # 

# 
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

# 3D
ax.bar3d(x3, y3, bottom, width, depth, height, shade=True, color='#9ddf6e')

# 
plt.savefig("3dbarplt_1.png")
plt.show()
```
