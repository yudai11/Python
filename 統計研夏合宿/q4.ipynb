## bound1とサンプリングの描画

```python
import numpy as np
import math
import matplotlib.pyplot as plt

# U[-1,1]に従う乱数を100つ生成してその合計を返す
def gen_rand_sum():
    arr_x = np.random.rand(100)
    sum = 0.0
    for i in range(100):
        arr_x[i] = (arr_x[i] - 0.5) * 2
        sum += arr_x[i]
    return sum

# Hoeffdingの不等式の上界を返す
def bound(t):
    ans = math.exp(- t * t / 200.0)
    return 2 * ans

# 経験確率の保存先
over_x = [0 for i in range(500)]
# 経験確率の保存
for k in range(10000):    
    rand_sum = gen_rand_sum()
    # i is x's index
    x = 0.01
    for i in range(500):
        if abs(rand_sum) > x:
            over_x[i] += 1
        x += 0.1        
for i in range(500):
    over_x[i] /= 10000.0
    
#plotに使用する配列の保存先
x_list =[]
y_list_1 = []
y_list_2 = []

#plotに使用する配列の保存
x = 0.01
for i in range(500):
    x_list.append(x)
    y_list_1.append(bound(x))
    y_list_2.append(over_x[i])
    x += 0.1
    
#plot
plt.plot(x_list, y_list_1, label = "Hoeffding's upper bound")
plt.plot(x_list, y_list_2, label = "empirical probability")     
plt.title("Hoeffding's inequality")
plt.xlabel("t")
plt.ylabel("probability")
plt.legend()
plt.show()


```
