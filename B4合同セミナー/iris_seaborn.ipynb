```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

urlprefix = 'https://vincentarelbundock.github.io/Rdatasets/csv/'
dataname = 'datasets/iris.csv'
iris = pd.read_csv(urlprefix + dataname)
setosa = iris.loc[iris['Species'] == 'setosa']
versicolor = iris.loc[iris['Species'] == 'versicolor']
virginica = iris.loc[iris['Species'] == 'virginica']
```

```python
sns.scatterplot(x='Petal.Length', y='Sepal.Length', data = setosa, color="blue", alpha = 0.5)
sns.scatterplot(x='Petal.Length', y='Sepal.Length', data = versicolor, color="green", alpha = 0.5)
sns.scatterplot(x='Petal.Length', y='Sepal.Length', data = virginica, color="red", alpha = 0.5)
plt.show()
```

```python
sns.kdeplot(setosa['Petal.Length'], fill = True, color="blue")
sns.kdeplot(versicolor['Petal.Length'], fill = True, color="green")
sns.kdeplot(virginica['Petal.Length'], fill = True, color="red")
plt.legend(['setosa','versicolor','virginica'])
plt.savefig('kde_isis.pdf')
plt.show()
```

```python
cols = ["blue","green","red"]
sns.scatterplot(x='Petal.Length', y='Sepal.Length', data=iris, hue='Species', color=cols)
plt.show()
```
