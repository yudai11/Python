```python
import numpy as np
import matplotlib.pyplot as plt

# データの作成
x = np.linspace(-6, 6, 1000)
y1 = (1 / np.sqrt(2 * np.pi)) * np.exp(-0.5 * x**2)
y2 = (1 / np.sqrt(2 * np.pi * 4)) * np.exp(-0.5 / 4 * x**2)

# プロットの作成
plt.rcParams['font.family'] = 'Yu Gothic'
plt.plot(x, y1, label='標準正規分布', color='blue')
plt.fill_between(x, y1, color='skyblue', alpha=0.4)
plt.plot(x, y2, label='標準正規分布', color='red')
plt.fill_between(x, y2, color='lightcoral', alpha=0.4)

# グラフの装飾
# plt.title('標準正規分布')
# plt.xlabel('x')
# plt.ylabel('確率密度')
# plt.legend()

# グリッドを表示
plt.grid(True)

# プロットの表示
plt.savefig("norm_dists2.png")
plt.show()
```
